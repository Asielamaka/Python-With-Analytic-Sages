## Question 6

```python
# Write a program that does the following:
# Create a CSV file called products.csv with the columns Product, Price and Quantity. 
# Add at least 4 rows of product data to the file using csv.writer().
# Read the file back using csv.DictReader() and print each row to the screen.
```

```python
import csv

with open ('products.csv', 'w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow (['Product', 'Price', 'Quantity'])
    writer.writerow (['Clove', 10, 20])
    writer.writerow (['Tumeric', 7, 30])
    writer.writerow (['Ginger', 20, 15])
    writer.writerow (['Thyme', 4, 12])
```

```python
with open ('products.csv', 'r') as file:
    read = csv.DictReader (file)
    for spices in read:
        print (spices)
```

```python
import numpy as np

scores = np.array([85, 92, 45, 60, 78, 50])

print ("Shape: ", scores.shape)
print ("Data type:", scores.dtype)

print ("Mean: ", np.mean(scores))
print ("Minimum: ", np.min(scores))
print ("Maximum: ", np.max(scores))
print ("Standard Deviation: ", np.std(scores))
```

```python
scores_above_60 = scores [scores >= 60]
print ("Scores from 60 and above: ", scores_above_60)
```

```python
import pandas as pd

students_data = {
    'Name': ['Ada', 'Obi', 'Jiibal', 'Tope', 'Feyi'],
    'Age': [20, 24, 23, 22, 20],
    'Score': [80, 53, 47, 65, 79],
    'Grade': ['A', 'B', 'B', 'A', 'B']
}

df = pd.DataFrame(students_data)
print(df.head(3))
```

```python
filtering_above_50 = df [df ['Score'] > 50]
print (filtering_above_50)
```

```python
sorted_df = df.sort_values (by = 'Score', ascending = False)
print (sorted_df)
```

```python
# This line of code will tell put True if a Score is >= 50, else it'll insert False.
df['Passed'] = np.where(df['Score'] >= 50, True, False)

# I'm converting my dataframe to a csv file
df.to_csv ('students_data.csv', index = False)

# This reloaded my csv file into a Dataframe on python
new_df = pd.read_csv('students_data.csv')
print (new_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading my downloaded USDT data as a DataFrame.
usdt_df = pd.read_csv('usdt-usd-max.csv')
usdt_df = usdt_df.head(5)

print (usdt_df)
```

```python
print ("Shape of Data: ", usdt_df.shape)

print ("Table Description: ", usdt_df.describe())
```

```python
plt.figure (figsize = (15, 3))

plt.plot(usdt_df['snapped_at'], usdt_df ['price'], color = 'green') 
plt.title ('USDT Chart')
plt.xlabel ('Date')
plt.ylabel ('Price')

plt.savefig ('line_chart.png')
plt.show
```

```python
plt.figure (figsize = (15, 4))

plt.bar (usdt_df ['snapped_at'], usdt_df ['market_cap'], color = 'blue')
plt.title ('USDT Bar Chart')
plt.xlabel ('Date')
plt.ylabel ('Market Cap')

plt.savefig ('bar_chart.png')
plt.show
```

```python
plt.pie (usdt_df['price'], labels = usdt_df ['snapped_at'], autopct = '%6.1f%%')

plt.title ('USDT Pie Chart')
plt.savefig('pie_chart.png')
plt.show
```
