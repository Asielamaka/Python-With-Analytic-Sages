## 🟢 Exercise 1: Crypto Price Fetcher

**Goal:** Practice GET requests + params + JSON parsing + error handling

### Task:

* Use the CoinGecko API:

```
https://api.coingecko.com/api/v3/simple/price
```

* Fetch prices for:

  * Bitcoin
  * Ethereum
* Print Status code check and error message if request fails
* Print output like:

```
Bitcoin: $XXXXX
Ethereum: $XXXXX

````python
import requests

import os
from dotenv import load_dotenv

load_dotenv()

Coingecko_API = os.getenv("Coingecko_API_Key")


url = "https://pro-api.coingecko.com/api/v3/simple/price"


headers = {"x-cg-pro-api-key": Coingecko_API}

params = {
    "names": "bitcoin,ethereum", 
    "vs_currencies": "usd"
}

response = requests.get(url, headers = headers, params=params)

data = response.json()

if response.status_code == 200:
    print (f"Status: Connected!")
else:
    print (f"Status: Error {response.status_code}")


print (f"Bitcoin: ${data['Bitcoin']['usd']}")
# print (f"Ethereum: ${data['Ethereum']['usd']}")


#print (data)
````

## 🟡 Exercise 2: User Input + API Call

**Goal:** Make dynamic API requests

### Task:

* Ask the user to input a cryptocurrency name (e.g., bitcoin)
* Fetch its price in USD
* Print:

```
The price of <coin> is $XXXXX
```

### Hint:

Use `input()` and pass the value into `params`

````python
import requests

import os
from dotenv import load_dotenv

load_dotenv ()

Coingecko_API = os.getenv ("Coingecko_API_Key")

url = "https://pro-api.coingecko.com/api/v3/simple/price"

headers = {"x-cg-pro-api-key": Coingecko_API}

token_name = input ("Enter the name of the token: ")

params =  {"ids": token_name,
           "vs_currencies": "usd"
           }

response = requests.get(url, headers = headers, params = params)

data = response.json()

print (f"The price of {token_name} is ${data[token_name]['usd']}")
````

## 🔴 Exercise 3: Mini API Script (Combine Everything)

**Goal:** Bring all concepts together

### Task:

Write a script that:

1. Asks user for a crypto name
2. Fetches price using API
3. Checks status code
4. Handles errors
5. Prints clean output

````python
import requests

import os
from dotenv import load_dotenv

load_dotenv()

Coingecko_API = os.getenv ("Coingecko_API_Key")

url = "https://pro-api.coingecko.com/api/v3/simple/price"

headers = {"x-cg-pro-api-key": Coingecko_API}

token_id = input ("Enter the name of the token").lower()

params = {"ids": token_id,
          "vs_currencies": "usd"
          }

response = requests.get (url, headers = headers, params = params)

data = response.json()

status = response.status_code

if status == 200:
    print (f"Status: {status}, Connected!")

else:
    print (f"Status: {status}")

print (f"The price of {token_id} is ${data[token_id]['usd']}")
````
